```python
from selenium import webdriver
import chromedriver_autoinstaller    #remove/comment this line in order to work with firefox and uncomment the last line
chromedriver_autoinstaller.install() #remove/comment this line in order to work with firefox and uncomment the last line.
driver = webdriver.Chrome()          #remove/comment this line in order to work with firefox and uncomment the last line.

#driver = webdriver.Firefox()        #uncomment this line in order to work with Firefox
```

```python
driver.get("https://www.twitter.com/twitter/with_replies") #Replace the given link to your twitter handle link
```

```python
import time
from selenium.webdriver.common.keys import Keys
x = 0
for i in range(5):       #Number of times to scroll the twitter profile page and extract tweets
    article = driver.find_elements_by_tag_name('article')
    
    for r in range(len(article)):
        print("\n\n"+article[r].text)
        print("__________________"+str(x)+" tweets loaded________________________")
        time.sleep(0.01)                #Delay after extracting one tweet  
        x = x+1
    art = driver.find_element_by_tag_name("article")
    art.send_keys(Keys.END)
    time.sleep(5)                       #Delay required till to scroll the page.
    
```
